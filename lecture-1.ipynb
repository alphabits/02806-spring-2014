# Exercise 2

In this exercise some basic Python operations are performed such as creating lists, defining functions, applying functions on lists, writing output to text files and working with strings.

First a list with the integers from 1 to 100 is created with the range function. The start parameter is included and the end parameter isn't so the call is:

```python
a = range(1, 101)
```

Next the numbers from 42 to 79 needs to be extracted from the list $a$. As with the range function the slice operation also includes the start index but the end index is exclusive. Also list indexes in Python is starting from 0, so to get numbers 42 to 79:

```python
b = a[41:79]
```

Next a function $f$ that multiplies the input $x$ with the input itself plus 2 is defined using the <code>def</code> keyword. Using a <code>for</code> loop, $f$ is applied to all elements of the list $b$

```python
def f(x):
    return x*(x+2)
```

```python
c1 = []
for x in b:
    c1.append(f(x))
```

A more pythonic way to apply a function to a list is to use a list comprehension

```python
c2 = [f(x) for x in b]
```

Independent of which method we use to create the list $c$, we now have an iterable object we want to save in a text file with one item per line. Using the <code>open</code> method we get a file object and write each element in $c$ to the file

```python
with open('lecture-1-list-c.txt', 'w') as fileobj:
    fileobj.write("\n".join(str(x) for x in c2))
```

Finally we need to work with strings and specifically the format operator <code>%</code>. The example from [Learn Python The Hard Way](http://learnpythonthehardway.org/book/ex6.html) is shown below

```python
x = "There are %d types of people." % 10
binary = "binary"
do_not = "don't"
y = "Those who know %s and those who %s." % (binary, do_not)

print x
print y

print "I said: %r." % x
print "I also said: '%s'." % y

hilarious = False
joke_evaluation = "Isn't that joke so funny?! %r"

print joke_evaluation % hilarious

w = "This is the left side of..."
e = "a string with a right side."

print w + e
```

```python
ys = [88, 56, 57, 56, 20, 39, 109, 73, 133, 79, 40, 35, 49, 43, 45, 
      33, 50, 66, 60, 60, 57, 35, 132, 217, 133, 135, 140, 100, 139, 
      167, 102, 69, 115, 90, 129, 86, 94, 72, 52, 92, 102, 157, 75, 
      120, 61, 49, 74, 96, 103, 89, 100, 64, 64, 152, 109, 122, 110, 
      97, 87, 69, 103, 114, 78]
xs = range(1, len(ys)+1)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(xs, ys)
plt.show()
```

```python
plt.plot(xs, ys, linestyle="dashed", marker="o", color="green")
plt.show()
```

```python
plt.plot(xs, ys, linestyle="dashed", marker="o", color="green")
_ = plt.ylim([10, 245])
_ = plt.yticks([10, 60, 110, 120, 140, 200])
_ = plt.xlim([-5, 67])
_ = plt.xticks([0, 10, 20, 30, 40, 50, 55, 60, 65])
```

```python
plt.plot(xs, ys, linestyle="dashed", marker="o", color="green")
_ = plt.xlim([-5, 68])
_ = plt.xlabel("The x-axis")
_ = plt.ylabel("The y-axis")
_ = plt.title("A nice plot")
```

```python
y_error = [round(0.1*y) for y in ys]
```

```python
plt.plot(xs, ys, linestyle="dashed", marker="o", color="green")
plt.errorbar(xs, ys, yerr=y_error, linestyle="None", marker="None", color="green")
_ = plt.xlim([-5, 68])
_ = plt.xlabel("The x-axis")
_ = plt.ylabel("The y-axis")
_ = plt.title("A nice plot")
```
